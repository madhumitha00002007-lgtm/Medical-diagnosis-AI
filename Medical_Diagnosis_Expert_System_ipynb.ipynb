```python
import pandas as pd

# 1. Healthcare Database Setup
medical_knowledge = {
    "Disease": [
        "Flu (Kaaichal)",
        "Food Poisoning",
        "Migraine",
        "Allergy",
        "Diabetes",
    ],
    "Core_Symptom": ["Fever", "Vomiting", "Headache", "Sneezing", "Thirst"],
    "Precaution": [
        "Take complete rest and drink warm water.",
        "Stay hydrated, drink ORS, and eat bland food.",
        "Rest in a dark room and avoid mobile/laptop screens.",
        "Avoid dust, smoke, and pets. Take antihistamines.",
        "Reduce sugar intake and check your blood glucose levels.",
    ],
}
df_medical = pd.DataFrame(medical_knowledge)


# 2. Interactive Diagnosis Engine
def run_diagnostic_system():
    print(
        "=================== PROJECT 2: AI MEDICAL DIAGNOSIS ==================="
    )
    print("Identify your primary symptom from the list:\n")

    symptoms = df_medical["Core_Symptom"].tolist()
    for i, symptom in enumerate(symptoms, 1):
        print(f" [{i}] {symptom}")
    print("=" * 70)

    try:
        user_choice = int(input("\nEnter the option number of your symptom: "))
        if 1 <= user_choice <= len(symptoms):
            selected = symptoms[user_choice - 1]
            report = df_medical[df_medical["Core_Symptom"] == selected].iloc[0]

            print("\n" + "═" * 50 + "\n 🩺 PRELIMINARY DIAGNOSIS REPORT\n" + "═" * 50)
            print(f"[*] Suspected Condition: {report['Disease']}")
            print(f"[*] Core Symptom Match : {report['Core_Symptom']}")
            print(f"[*] Recommended Care   : {report['Precaution']}")
            print("═" * 50)
        else:
            print("[!] Invalid option selected.")
    except ValueError:
        print("[!] Input Error. Please enter a valid index number.")


run_diagnostic_system()
```
